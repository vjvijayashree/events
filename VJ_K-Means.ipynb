```python
import pandas as pd
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
#assign column names
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
dataset = pd.read_csv(url, names=names)
dataset.head()
#obtain values for the clustering algorithm
data=dataset.iloc[:,[0,1,2,3]].values
data[:,0]
#plot the graph
import matplotlib.pyplot as plt
#plotforsample data
plt.scatter(data[:,0], data[:,1], c='k')
plt.title('Scatter-plot of IRIS')
plt.show()
#k-means 
plt.scatter(data[:,2], data[:,3], c='k')
plt.title('k-means of Scatter-plot of petal comparison')
#plt.set_ylabel('Y-Axis')
#plt.set_xlabel('X-Axis')
plt.show()
```

```python
#k-means 
plt.scatter(data[:,2], data[:,3], c='red')
plt.title('K-means of Scatter-plot of petal comparison')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
#Apply_machineLearning
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
#apply k-means, random state is used for consistency 
kmeans= KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=3) 
#obtain predicted values
pred=kmeans.fit_predict(data)
pred
#check clasess from dataset
dataset['Class'].values
#changing classes into numerical values
actual=[]
for i in dataset['Class']:
    if i =='Iris-setosa':
        actual.append(1)
    elif i =='Iris-versicolor':
        actual.append(0)
    else:
        actual.append(2)
#calculate accuracy of model using sklearn function
accuracy_score(actual,pred)
len(dataset)

#draw prediction graph
pred==0
data[:,0]
```

```python
w = []
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10 , random_state =3)
    kmeans.fit(data)
    w.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.plot(range(1,11), w,marker = 'o')
plt.title('Elbow graph')
plt.xlabel('No. of Clusters')
plt.ylabel('Inertia')
plt.show()
```
